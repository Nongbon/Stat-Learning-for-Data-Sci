*Time Resampling การหาผลสรุปเป็นรายปี*

resample() function:

W:weekly frequency
M:month end frequency
Q:quarter end frequency
A:year end frequency

```
df =   pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/a10.csv', parse_dates=['date'])
df.set_index('date',inplace=True)
df1 =df.resample(rule='A').mean()
df1.reset_index(inplace=True)
print(df1)
```

จงสร้างตัวแบบพยากรณ์ข้อมูลด้านบนโดยใช้ตัวแบบการถดถอย

```
df2=range(18)
df2=pd.DataFrame({'t':df2})
#print(df2)
ts1 = pd.concat([df1,df2],axis=1)
#print(ts1)
```

สร้าง Linear regression model

```
from scipy import stats
# StatsModels
import statsmodels.api as sm
import statsmodels.formula.api as smf
model = smf.ols('value ~ t', ts1)
results = model.fit()
print(results.summary())
```

```
yhat = model.predict(results.params)

print(yhat)
```

พล๊อตกราฟ y กับ  yhat

```
# Plot y and yhat together
from datetime import datetime
from matplotlib import dates as mpl_dates


y=ts1['value']

fig, ax = plt.subplots(figsize=(12,5), dpi=100)
ax.plot(y, marker='.', linestyle='-', linewidth=0.5, label='actual')
ax.plot(yhat,marker='o', markersize=8, linestyle='-', label='forecast')
ax.set_ylabel('Sales')
plt.xticks(ts1.t)
ax.legend();
```

```
df =   pd.read_csv("data.csv").reset_index().rename(columns={'index': 't', 'Sales (GWh) ': 'Sales (GWh)'})
df
```

```
model = smf.ols('Q("Sales (GWh)") ~ t', df)
result = model.fit()
print(result.summary())
```

```
result.params
```

```
yhat = model.predict(result.params)

print(yhat)
```

```
y = df['Sales (GWh)']

fig, ax = plt.subplots(figsize=(12,5), dpi=100)
ax.plot(y, marker='.', linestyle='-', linewidth=0.5, label='actual')
ax.plot(yhat, marker='o', markersize=8, linestyle='-', label='forecast')
ax.set_ylabel('Sales (GWh)')
plt.xticks(df.t)
ax.legend();
```
